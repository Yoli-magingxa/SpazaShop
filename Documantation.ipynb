**Comparisons of THREE models:**

Prophet appears to be the most suitable model for predicting sales in the spaza shop, primarily due to its ability to account for holidays and events, which seem to be important factors in the shop's sales trends.
Prophet is designed to handle time series data with seasonalities, holidays, and non-linear trends. It is particularly useful when there are known events like public holidays or special promotions that can significantly impact sales. Prophet can automatically incorporate these into the forecast.

**ARIMA**

ARIMA is effective for modeling linear trends and seasonality, especially when the data is stationary. It excels in short-term forecasting where patterns like autoregression (past values affecting current values) and moving averages are strong.It assumes the data is stationary and cannot handle complex seasonality or holidays as efficiently. There are non-linear patterns or/and multiple factors affecting sales (like promotions, events, or holidays), therefore ARIMA might not capture them effectively.


**LSTM (Long Short-Term Memory Neural Network)**


Pros: LSTM is powerful for handling non-linear patterns, sequential data, and long-term dependencies. It excels in capturing complex interactions between different variables, such as sales, inventory, promotions, and events.
LSTM requires more data to train effectively and is computationally more intensive. It is also prone to overfitting if not tuned properly.
